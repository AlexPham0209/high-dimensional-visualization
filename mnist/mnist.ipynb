## Importing Libraries

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import umap
import seaborn as sns
import numpy as np
import pandas as pd
```

## Loading MNIST Digit Dataset

```python
(x, y), _ = tf.keras.datasets.mnist.load_data()
print(f"Shape of x: {x.shape}")
print(f"Shape of y: {y.shape}")

plt.imshow(x[0])
plt.show()
```

## Standardizing Images

```python
x = x.reshape(x.shape[0], -1)
# x = StandardScaler().fit_transform(x)

# plt.imshow(x[0].reshape(28, 28))
# plt.show()
```

## Principal Component Analysis

```python
x_tranformed = PCA(n_components=2).fit_transform(x)
df = pd.DataFrame({"x": x_tranformed[:2000, 0], "y": x_tranformed[:2000, 1], "label": y[:2000]})
sns.scatterplot(x="x", y="y", hue="label", palette="tab10", data=df)
plt.show()
```

## TSNE

```python
x_tranformed = TSNE(n_components=2).fit_transform(x[:2000])
df = pd.DataFrame({"x": x_tranformed[:2000, 0], "y": x_tranformed[:2000, 1], "label": y[:2000]})
sns.scatterplot(x="x", y="y", hue="label", palette="tab10", data=df)
plt.show()
```

```python
x_tranformed = umap.UMAP(n_components=2).fit_transform(x[:2000])
df = pd.DataFrame({"x": x_tranformed[:2000, 0], "y": x_tranformed[:2000, 1], "label": y[:2000]})
sns.scatterplot(x="x", y="y", hue="label", palette="tab10", data=df)
plt.show()
```
